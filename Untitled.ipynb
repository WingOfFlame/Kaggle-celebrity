```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
from collections import Counter
%matplotlib inline
```

## Load Images

```python
images = []
filenames = []
celeberty_names = []
for filename in glob.glob('input/images-train/*/*.jpg'): 
    im=mpimg.imread(filename)
    images.append(im)
    filenames.append(filename)
    celeberty_names.append(filename.split('/')[2])
```

### Look at a sample

```python
print(f'Total files: {len(filenames)}')
print(f'Total celeberties: {len(Counter(celeberty_names)))}')
print(filenames[1])
print(celeberty_names[1])
plt.imshow(images[1]) 
```

### Convert to numpy array

```python
#TODO: images dimension not the sae
#images = np.stack(images) 
filenames = np.array(filenames)
celeberty_names = np.array(celeberty_names)
```
